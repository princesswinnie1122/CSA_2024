# 1. (90%) Bisection method (二分法求根)
The bisection method is a numerical root-finding algorithm that works by repeatedly dividing an interval in half and determining which half of the interval contains a root of a given function.

Suppose we have a continuous function f(x) on the interval [a, b] and we want to find a root (i.e., a value of x such that f(x) = 0) in this interval. The bisection method starts by computing the value of f(a) and f(b) to determine if the root is contained in the interval [a, b]. If f(a) and f(b) have opposite signs, then by the intermediate value theorem, there must be at least one root in the interval. 

Next, the bisection method bisects the interval by computing the midpoint c = (a + b)/2. Depending on the sign of f(c), the method selects either the interval [a, c] or [c, b] for the next iteration. Specifically, if f(a) and f(c) have opposite signs, then the root is in the interval [a, c], so the method sets b = c and repeats. Otherwise, if f(c) and f(b) have opposite signs, then the root is in the interval [c, b], so the method sets a = c and repeats. The process is repeated until the interval becomes sufficiently small or until a desired level of accuracy is reached. This procedure is illustrated as the figure below.

The bisection method is guaranteed to converge to a root if f(x) is a continuous function on the interval [a, b] and if f(a) and f(b) have opposite signs. However, the convergence rate of the bisection method is relatively slow compared to other root-finding algorithms, such as the Newton-Raphson method, especially if the interval is large and the function is slowly varying.

In this question, we will implement the bisection method in python to help us find roots for a non-linear function:

````python
import numpy as np

def f(x):
    return np.sin(x)

def root_finder(f, a, b, tol):
    n = 0
    while True:
        m = f(a) * f(b)
        if m > 0:
            return None, None
        elif m == 0:
            if abs(f(a)) < tol:
                return a, n
            elif abs(f(b)) < tol:
                return b, n
        else:
            c = (a + b) / 2
            if abs(f(c)) < tol:
                return c, n
            else: 
                if f(c) * f(a) < 0:
                    b = c
                else:
                    a = c
                n+=1
            
for tol in np.logspace(-1, -8, 8):
    root, count = root_finder(f, 3, 4, tol)
    error = ((root - np.pi) / np.pi) * 100
    print(f"Estimated root: {root:.16f}; relative error: {error:10.2e}% after {count-1:4} iterations")
    #count-1: 不考慮當回合
````
